#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
colunas = ['tipo_renda','educacao','estado_civil','tipo_residencia']
demo_encoded = pd.get_dummies(demo,columns=colunas)
demo_encoded
````

````python
demo_encoded['sexo'] = demo_encoded['sexo'].map({'M':0,'F':1})
demo_encoded['posse_de_veiculo'] = demo_encoded['posse_de_veiculo'].map({'N':0,'Y':1})
demo_encoded['posse_de_imovel'] = demo_encoded['posse_de_imovel'].map({'N':0,'Y':1})
````

````python
demo_encoded['mau'] = demo_encoded['mau'].astype(int)
````
